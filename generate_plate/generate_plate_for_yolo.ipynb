```python
import os
import arabic_reshaper
import cv2
from PIL import ImageFont, ImageDraw, Image
import numpy as np
```

```python

from bidi.algorithm import get_display

def persian(text):
    return get_display(arabic_reshaper.reshape(text))
```

```python
# w = 50

org=[] 
org.append([60  , 40])
org.append([120 , 40])
org.append([180 , 25])
org.append([270 , 40])
org.append([330 , 40])
org.append([390 , 40])
org.append([470 , 60])
org.append([525 , 60])

number_of_plates = 1000
for n in range(number_of_plates):
    label_file = ""
    img = cv2.imread("./template/omomi.png")
    image_width = img.shape[1]
    image_height = img.shape[0]
    # print(image_width)
    # print(image_height)
    im = Image.fromarray(img)
    label_file = ""
    for i in range(8):
        numbers = [
            {
                "sign": "0",
                "id": 0
            },
            {
                "sign": "1",
                "id": 1
            },
            {
                "sign": "2",
                "id": 2
            },
            {
                "sign": "3",
                "id": 3
            },
            {
                "sign": "4",
                "id": 4
            },
            {
                "sign": "5",
                "id": 5
            },
            {
                "sign": "6",
                "id": 6
            },
            {
                "sign": "7",
                "id": 7
            },
            {
                "sign": "8",
                "id": 8
            },
            {
                "sign": "9",
                "id": 9
            },]
        letters = [
            {
                "sign": "الف",
                "id": 10
            },
            {
                "sign": "ب",
                "id": 11
            },
            {
                "sign": "پ",
                "id": 12
            },
            {
                "sign": "ت",
                "id": 13
            },
            {
                "sign": "ث",
                "id": 14
            },
            {
                "sign": "ج",
                "id": 15
            },
            {
                "sign": "چ",
                "id": 16
            },
            {
                "sign": "ح",
                "id": 17
            },
            {
                "sign": "خ",
                "id": 18
            },
            {
                "sign": "د",
                "id": 19
            },
            {
                "sign": "ذ",
                "id": 20
            },
            {
                "sign": "ر",
                "id": 21
            },
            {
                "sign": "ز",
                "id": 22
            },
            {
                "sign": "ژ",
                "id": 23
            },
            {
                "sign": "ص",
                "id": 24
            },
            {
                "sign": "ض",
                "id": 25
            },
            {
                "sign": "ط",
                "id": 26
            },
            {
                "sign": "ظ",
                "id": 27
            },
            {
                "sign": "ع",
                "id": 28
            },
            {
                "sign": "غ",
                "id": 29
            },
            {
                "sign": "ف",
                "id": 30
            },
            {
                "sign": "ق",
                "id": 31
            },
            {
                "sign": "ک",
                "id": 32
            },
            {
                "sign": "گ",
                "id": 33
            },
            {
                "sign": "ل",
                "id": 34
            },
            {
                "sign": "م",
                "id": 35
            },
            {
                "sign": "ن",
                "id": 36
            },
            {
                "sign": "و",
                "id": 37
            },
            {
                "sign": "ه",
                "id": 38
            },
            {
                "sign": "ی",
                "id": 39
            },
        ]
        
        letter = np.random.choice(numbers,  size=(1))
        if i==2:
            # letter = np.random.choice(["الف","ب","ج", "د", "ر", "ع"],  size=(1))
            letter = np.random.choice(letters,  size=(1))

        # letter = np.random.choice(["3"],  size=(1))
        
        position = org[i]
        # with Image.open("./template/omomi.png") as im:

        
        # img = Image.fromarray(im)
        # img = im
        draw = ImageDraw.Draw(im)
        text = letter[0]["sign"]
        if i < 6 and i!=2:
            size=80
        if i == 2:
            size= 65
        if i>=6:
            size=70

        font = ImageFont.truetype("./fonts/roya_bold.ttf", size=size)
        draw.text(position, persian(text), fill=(0,0,0), font=font)
        
        # if i!=2:
        #     class_id = int(letter)
        # else:
        class_id = letter[0]["id"] 
        
        if i!=0:
            label_file = label_file + '\n'


        if i < 6 and i!=2:
            x1 = position[0]
            y1 = position[1]
            x2 = position[0] + 50
            y2 = position[1] + 60


        if i==2:
            x1 = position[0]
            y1 = position[1] 
            x2 = position[0] + 80
            y2 = position[1] + 90
        
        if i>=6:
            x1 = position[0]
            y1 = position[1]
            x2 = position[0] + 45
            y2 = position[1] + 50

        # draw.rectangle([(x1, y1) , (x2, y2)], outline ="black")
        
        xc = ( x1 + x2 ) / 2
        yc = ( y1 + y2 ) / 2
        w = x2 - x1
        h = y2 - y1

        xc = xc/image_width
        yc= yc/image_height
        w = w/image_width
        h = h/image_height
        # label_file = label_file + f"{class_id} {x1} {y1} {w} {h}"
        label_file = label_file + f"{class_id} {xc} {yc} {w} {h}"
    
    img = np.array(im)
    cv2.imwrite(f'./generated/temp{n}.png' ,img)
    with open(f'./generated/temp{n}.txt', 'w')as f:
        f.write(label_file)
        # im.show()
```
